 Visual Analysis - Univariate, Bivariate after pre-processing of the data

```python
# Note: Pre-processing involves unnesting of the data in columns like Actor, Director, Country
```

```python
# STEPS FOR PRE-PROCESSING DATA
```

```python
# Preprocessed data
import pandas as pd

# Load the dataset
df = pd.read_csv('netflix.csv')

# Unnesting columns
df['cast'] = df['cast'].str.split(', ')
df['director'] = df['director'].str.split(', ')
df['country'] = df['country'].str.split(', ')
df['listed_in'] = df['listed_in'].str.split(', ')

# Explode unnested columns
df = df.explode('cast')
df = df.explode('director')
df = df.explode('country')
df = df.explode('listed_in')

# Reset index
df = df.reset_index(drop=True)

# Drop missing values
df = df.dropna()

# Convert release year to integer
df['release_year'] = df['release_year'].astype(int)

# Convert duration column to numeric
# Extract only the number (e.g. "90 min" → 90, "2 Seasons" → 2)
df['duration_numeric'] = df['duration'].str.extract('(\d+)').astype(float)

# Remove duplicates
df = df.drop_duplicates()

# Save clean dataset
df.to_csv('preprocessed_netflix.csv', index=False)
```

Missing Value & Outlier check

1. Missing Value

```python
# Check for missing values
missing_values = df.isnull().sum()

# Display the count of missing values for each attribute
print(missing_values)
```

```python
missing_count = df.isnull().sum()
missing_percent = (missing_count / len(df)) * 100

missing_report = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing %': missing_percent.round(2)
})

print(missing_report)
```

2. Outliers

```python
# Using IQR method
Q1 = df['duration_numeric'].quantile(0.25)
Q3 = df['duration_numeric'].quantile(0.75)
IQR = Q3 - Q1

# Filter outliers
df_no_outliers = df[(df['duration_numeric'] >= Q1 - 1.5*IQR) &
                    (df['duration_numeric'] <= Q3 + 1.5*IQR)]
```

```python
print(df_no_outliers.head())
```

```python
# Check how many rows remain after removing outliers
print("Original rows:", len(df))
print("Rows without outliers:", len(df_no_outliers))
```
