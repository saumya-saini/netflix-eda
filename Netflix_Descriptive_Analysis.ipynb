```python
# Preprocessing
import pandas as pd
df = pd.read_csv("netflix.csv")

df['cast'] = df['cast'].str.split(', ')
df['director'] = df['director'].str.split(', ')
df['country'] = df['country'].str.split(', ')
df['listed_in'] = df['listed_in'].str.split(', ')

df = df.explode('cast')
df = df.explode('director')
df = df.explode('country')
df = df.explode('listed_in')

df = df.reset_index(drop = True)

df = df.dropna()

df['release_year'] = df['release_year'].astype(int)

#Feature Engineering
df['duration_numeric'] = df['duration'].str.extract('(\d+)').astype(float)
df['movie_minutes'] = df.apply(lambda x: x['duration_numeric'] if x['type'] == "Movie" else None, axis = 1)
df['tv_seasons'] = df.apply(lambda x: x['duration_numeric'] if x['type'] == "TV Show" else None, axis = 1)

df = df.drop(columns = ['duration_numeric'])
df = df.drop_duplicates()

df.to_csv('preprocessed_netflix.csv', index = False)
```

Business Insights
1. Movies dominate the Netflix catalog

The analysis shows that movies make up the majority of Netflix’s content library. This indicates that the platform historically relied heavily on films to build its catalog.

Inference:
Movies help Netflix maintain a large content library and attract viewers looking for standalone entertainment.

2. TV shows have increased significantly in recent years

While movies still dominate overall, the number of TV shows added has increased rapidly in recent years.

Inference:
Netflix appears to be investing more in series content to keep viewers engaged for longer periods.

3. Content production is concentrated in a few countries

The majority of Netflix content comes from countries such as the United States, India, and the United Kingdom.

Inference:
These countries play a major role in global content production and contribute significantly to Netflix’s catalog.

4. Certain actors and directors appear frequently

Some actors and directors appear multiple times in the dataset.

Inference:
Netflix likely collaborates repeatedly with certain performers and filmmakers who have proven to attract audiences.

5. Movie releases have grown significantly after 2010

The number of movies released each year increased sharply after 2010.

Inference:
The growth of streaming services likely increased demand for digital content production.

6. Certain months have higher TV show releases

The data shows that some months (for example February and August) have higher numbers of TV show releases.

Inference:
Netflix may strategically schedule releases during periods when viewers are more active.

7. Global diversity in content

The dataset includes content from many different countries and cultures.

Inference:
Netflix aims to serve an international audience by offering diverse content.

Recommendations
1. Increase investment in TV series

TV shows keep viewers engaged for longer periods. Increasing the production of high-quality series can help maintain subscriber retention.

2. Expand content production in emerging markets

Netflix should increase content production in countries with growing streaming audiences such as regions in Asia, Africa, and Latin America.

3. Continue collaborating with popular actors and directors

Frequent collaborations with well-known actors and directors can attract viewers and increase engagement.

4. Release new content during high-activity months

Since certain months show higher release activity, Netflix should strategically launch major titles during these periods to maximize audience attention.

5. Focus on regional storytelling

Producing content that reflects local cultures and languages can help Netflix expand its international subscriber base.

6. Maintain a balance between movies and TV shows

While movies attract viewers quickly, TV shows keep them engaged longer. Maintaining a balanced mix of both formats can help improve viewer satisfaction.

7. Invest in original content

Original content differentiates Netflix from competitors and strengthens its brand identity.
